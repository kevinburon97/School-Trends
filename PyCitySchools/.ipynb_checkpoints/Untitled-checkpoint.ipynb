```python
import pandas as pd
file_schools = "Resources/schools_complete.csv"
file_students = "Resources/students_complete.csv"

schools = pd.read_csv(file_schools)
students = pd.read_csv(file_students)
```

```python
school_data = pd.merge(students, schools, on=["school_name"])
```

```python
Total_Schools_dist = [len(schools["School ID"])]
Total_Students_dist = sum(schools["size"]),
Total_Budget_dist = sum(schools["budget"]),
Avg_Math_Score_dist = sum(students["math_score"])/len(students["math_score"]),                        
Avg_Read_Score_dist = sum(students["reading_score"])/len(students["reading_score"]),
Perc_passing_Math_dist = 100*(len(Pass_math["P/F Math"])/len(school_data)),
Perc_passing_Reading_dist = 100*(len(Pass_read["P/F Read"])/len(school_data)),
Overall_Passing_dist = [100*(len(total_pass["Student ID"])/len(school_data))]
```

```python
bins = [1,69,100]
Group = ["F","P"]
school_data["P/F Math"] = pd.cut(school_data["math_score"], bins, labels = Group)   
school_data["P/F Read"] = pd.cut(school_data["reading_score"], bins, labels = Group)

```

```python
#Find the number of P 
Pass_math = school_data.loc[school_data["P/F Math"] == "P"]
Pass_read = school_data.loc[school_data["P/F Read"] == "P"]
total_pass = pd.merge(Pass_math, Pass_read, on="Student ID")
```

```python
DistSummary = pd.DataFrame({
    "Total Schools:": Total_Schools_dist,
    "Total Students":Total_Students_dist,
    "Total Budget": Total_Budget_dist,
    "Avg Math Score":Avg_Math_Score_dist,                        
    "Avg Read Score":Avg_Read_Score_dist,
    "% Passing Math" :Perc_passing_Math_dist,
    "% Passing Reading" :Perc_passing_Reading_dist,
    "Overall Passing" : Overall_Passing_dist
                     })
DistSummary
```

```python
#total students
Total_students = school_data["school_name"].value_counts()
```

```python
#total budget
Total_budget = school_data.groupby(["school_name"])["budget"].mean()
```

```python
#per student budget
Per_stu_budget = Total_budget/Total_students
```
